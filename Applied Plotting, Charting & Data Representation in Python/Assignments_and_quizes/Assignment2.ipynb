```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(25,'391a2922ad597ba080f4b99dea6d62842562d64845ef5df1a384561e')
```

```python
import pandas as pd
import numpy as np

raw = pd.read_csv('data/C2A2_data/BinnedCsvs_d25/391a2922ad597ba080f4b99dea6d62842562d64845ef5df1a384561e.csv')
raw.head()
raw['Year'] = raw['Date'].str[:4]
tmax_05_14 = raw.where((raw['Element'] == 'TMAX') & (raw['Year'] != '2015') & (raw['Date'].str[5:] != '02-29')).dropna()
tmax_05_14 = tmax_05_14.drop(['Element'], axis = 1).sort(['Date', 'ID'])
tmax_05_14['Date'] = tmax_05_14['Date'].str[5:]
tmax_05_14 = tmax_05_14.groupby('Date').agg({'Data_Value' : np.max}).reset_index()
tmax_05_14.head()
```

```python
tmin_05_14 = raw.where((raw['Element'] == 'TMIN') & (raw['Year'] != '2015') & (raw['Date'].str[5:] != '02-29')).dropna()
tmin_05_14 = tmin_05_14.drop(['Element'], axis = 1).sort(['Date', 'ID'])
tmin_05_14['Date'] = tmin_05_14['Date'].str[5:]
tmin_05_14 = tmin_05_14.groupby('Date').agg({'Data_Value' : np.min}).reset_index()
tmin_05_14.head()
```

```python
tmax_15 = raw.where((raw['Element'] == 'TMAX') & (raw['Year'] == '2015') & (raw['Date'].str[5:] != '02-29')).dropna()
tmax_15 = tmax_15.drop(['Element'], axis = 1).sort(['Date', 'ID'])
tmax_15['Date'] = tmax_15['Date'].str[5:]
tmax_15 = tmax_15.groupby('Date').agg({'Data_Value' : np.max}).reset_index()
tmax_15.head()
```

```python
tmin_15 = raw.where((raw['Element'] == 'TMIN') & (raw['Year'] == '2015') & (raw['Date'].str[5:] != '02-29')).dropna()
tmin_15 = tmin_15.drop(['Element'], axis = 1).sort(['Date', 'ID'])
tmin_15['Date'] = tmin_15['Date'].str[5:]
tmin_15 = tmin_15.groupby('Date').agg({'Data_Value' : np.min}).reset_index()
tmin_15.head()
```

```python
# find dates where record of last 10 years was broken max and min
record_broken_max = np.where(tmax_15['Data_Value'] > tmax_05_14['Data_Value'])[0]
record_broken_min = np.where(tmin_15['Data_Value'] < tmin_05_14['Data_Value'])[0]
```

```python
import matplotlib.pyplot as plt

# size and details of plot
plt.figure(figsize=(15,7))
plt.title('Record high and record low temperatures by day of the year')
plt.ylabel('Temperature (tenths of degrees C)')
plt.xlabel('Date in MM-DD format (e.g. 01-24 = January 24)')

# y ticks and xticks
plt.yticks(np.linspace(tmin_05_14['Data_Value'].min()-100, tmax_05_14['Data_Value'].max()+100, 10))
plt.xticks(range(len(tmin_05_14['Data_Value']))[::20], tmin_05_14['Date'][::20], size='small', rotation = 45)

# plotting Maximum temperature and minimum temperature datewize
plt.plot(tmax_05_14.Data_Value, 'm', label = 'Maximum Temperature by day of year(2005-2014)')
plt.plot(tmin_05_14.Data_Value, 'y', label = 'Minimum Temperature by day of year(2005-2014)')

# plotting record broken max and min
plt.scatter(record_broken_max, tmax_15.iloc[record_broken_max]['Data_Value'],\
            s = 10, c = 'k', alpha = 1, label = 'Maximum temperature record broken in 2015')
plt.scatter(record_broken_min, tmin_15.iloc[record_broken_min]['Data_Value'],\
            s = 10, c = 'b', alpha = 1, label = 'Minimum temperature record broken in 2015')

# shading the area under two plots
plt.gca().fill_between(range(len(tmin_05_14)), tmax_05_14['Data_Value'], tmin_05_14['Data_Value'],\
                       color = 'cyan', alpha = 0.2, label = 'Area b/w Max and Min temperatures')

# add legend to the plot
plt.legend(loc = 8, frameon = False)

# show
plt.show()
```
